###데이터 전처리-모델 특성 고려

```python
import pandas as pd

# 결측치 확인
missing_summary = df.isnull().sum()
print("결측치 개수:\n", missing_summary[missing_summary > 0])

# --- 결측치 처리 전략 예시 ---

# 1. 수치형 변수는 평균으로 대체
num_cols = ['DURATION', 'AMOUNT', 'AGE', 'INSTALL_RATE', 'NUM_CREDITS', 'NUM_DEPENDENTS']
for col in num_cols:
    if col in df.columns and df[col].isnull().sum() > 0:
        df[col].fillna(df[col].mean(), inplace=True)

# 2. 범주형 변수는 최빈값으로 대체
cat_cols = ['CHK_ACCT', 'HISTORY', 'SAV_ACCT', 'EMPLOYMENT', 'PRESENT_RESIDENT', 'JOB']
for col in cat_cols:
    if col in df.columns and df[col].isnull().sum() > 0:
        df[col].fillna(df[col].mode()[0], inplace=True)

# 3. 이진 변수도 결측 있으면 0으로 대체 (가정: 없음 의미)
binary_cols = ['NEW_CAR', 'USED_CAR', 'FURNITURE', 'RADIO/TV', 'RETRAINING', 'EDUCATION',
               'MALE_DIV', 'MALE_SINGLE', 'MALE_MAR_WID', 'CO_APPLICANT', 'GUARANTOR',
               'REAL_ESTATE', 'PROP_UNKN_NONE', 'OTHER_INSTALL', 'RENT', 'OWN_RES',
               'TELEPHONE', 'FOREIGN']
for col in binary_cols:
    if col in df.columns and df[col].isnull().sum() > 0:
        df[col].fillna(0, inplace=True)

# 결측치 처리 결과 확인
print("남은 결측치:\n", df.isnull().sum().sum())
```

```python
import numpy as np

# 수치형 변수 목록
numeric_cols = ['DURATION', 'AMOUNT', 'AGE', 'INSTALL_RATE', 'NUM_CREDITS', 'NUM_DEPENDENTS']

# 이상치 처리: IQR 기반 Clipping
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[col] = np.clip(df[col], lower_bound, upper_bound)

# 로그 변환 적용 (양의 값만 있는 변수에 대해)
log_cols = ['AMOUNT', 'DURATION']
for col in log_cols:
    df[col + '_log'] = np.log1p(df[col])  # log1p: log(1 + x)

# 결과 확인
print(df[[col for col in df.columns if 'log' in col]].head())
```

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor

# X: 다중선형회귀 독립변수 데이터프레임 (스케일링/인코딩 후)

def calculate_vif(X):
    vif_data = pd.DataFrame()
    vif_data["feature"] = X.columns
    vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))]
    return vif_data

vif = calculate_vif(X)
print(vif)

# 일반적으로 VIF 5 이상(또는 10 이상)인 변수는 다중공선성 의심
# 높은 VIF 변수 제거 후 재계산 반복 가능
```
